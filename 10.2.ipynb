```python
def f(x, y):
    z = x**2 + y**2
    return z
```

```python
x = np.linspace(-2,2,20)
y = np.linspace(-4,4,30)

X,Y = np.meshgrid(x,y)
```

```python
X.shape
Y.shape
```

```python
Z = f(X, Y)
```

```python
Z.shape
```

```python
plt.contour(X, Y, Z, 20, cmap="RdGy")
plt.colorbar()
```

```python
plt.contourf(X, Y, Z, 100, cmap="RdGy")
plt.colorbar()
```

```python
plt.imshow(Z, extent = [-2,2,-4,4], origin = "lower", cmap="RdGy")
plt.colorbar()
plt.axis(aspect="image")
```

```python
plt.imshow(Z, extent = [-2,2,-4,4], origin = "lower", cmap="RdGy")
plt.colorbar()
```

```python
# two plots at the same time
```

```python
contours = plt.contour(X, Y, Z, 3)
plt.clabel(contours, inline=True, fontsize=20)
plt.imshow(Z, extent=[-2, 2, -4, 4], origin="tower")
plt.colorbar()
```
